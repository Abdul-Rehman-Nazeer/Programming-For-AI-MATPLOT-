```python
# Q1.a
import matplotlib.pyplot as plt

names= [
    "Ali","Sara","Usman","Ayesha",
    "Ahmed","Hina","Bilal","Zainab","Omar",
    "Sadia","Imran","Aisha","Saad","Saima",
    "Tariq","Nida","Fahad","Samina","Yasir","Huma"
]

heights = [
    110, 175, 90, 180, 155,
    130, 70, 180, 210, 230,
    110, 175, 160, 145, 180,
    155, 165, 170, 175, 180
]

c=['green','blue','red','aqua','violet','black','pink','magenta','grey','orange','darkgreen','purple','maroon','gold','cyan']

plt.xlabel('Students')
plt.ylabel('Height (cm)')
plt.title('Student Height Data') 
plt.bar(names,heights,color=c)
plt.xticks(rotation=45)
plt.show()
```

```python
# Q.3You have a list of ice cream flavors and the number of scoops sold for each flavor. Create a pie
# chart to show the distribution of ice cream sales.

import matplotlib.pyplot as plt

flavors = ['Blueberry', 'Chocolate', 'Strawberry', 'Mint', 'Caramel']
col=['blue','brown','pink','grey','yellow']
scoops = [300, 400, 200, 350, 150]

explode = (0, 0.1, 0, 0, 0)#showing the most popullar flavoor

plt.pie(scoops, labels=flavors, explode=explode, autopct='%1.1f%%',colors=col)

plt.title('Distribution of Ice Cream Sales')

plt.show()
```

```python
#q.4

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students_info.csv")

age_groups = ['18-25', '26-30', '31-35', '36 and above']#use for labelling

bins = [18, 26, 31, df["Age"].max()]

df['Age Group'] = pd.cut(df["Age"], bins, labels=age_groups)

age_distribution = df['Age Group'].value_counts()

plt.pie(age_distribution, labels=age_distribution.index, autopct='%1.1f%%', startangle=140)

plt.axis('equal')
plt.title('Distribution of Student Ages')
plt.show()
```

```python
# q.5You're interested in the gender distribution among the students in your dataset. Create a pie
# chart to visualize the percentage of male and female students.


import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("students_info.csv")
lab=["MALE","FEMALE"]
gender = data['Gender'].value_counts()

plt.pie(gender, labels=lab, autopct='%1.1f%%', colors=['blue', 'pink'])

plt.title('Gender Distribution Among Students')
plt.show()
```

```python
# q.6 Write a Python program to draw a scatter plot comparing two subject marks of Mathematics
# and Science. Use marks of 10 students and also add legends for better understanding of the
# data.

import matplotlib.pyplot as plt

math = [92, 55, 88, 65, 100, 75, 80, 78, 94, 60]
science = [78, 65, 92, 55, 97, 70, 85, 90, 82, 75]

students = ['Student 1', 'Student 2', 'Student 3', 'Student 4', 'Student 5',
            'Student 6', 'Student 7', 'Student 8', 'Student 9', 'Student 10']

plt.scatter(math, science, label='Students', s=100)


for i in range(len(math)):
    plt.text(math[i], science[i], students[i])

plt.xlabel('Mathematics Marks')
plt.ylabel('Science Marks')
plt.title('Mathematics vs Science Marks of Students')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# q.7Write a Python program to draw a scatter plot to find sea level rise in past 100 years. Use the
# provided dataset.

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('epa-sea-level.csv')

data = data[data['Year'] >= 1920]

plt.scatter(data['Year'], data['CSIRO Adjusted Sea Level'], label='Sea Level', marker='o', color='blue')

plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Global Mean Sea Level Rise')
plt.legend()
plt.grid(True)

plt.show()
```
